### Ejercicio 3
Crea una matriz bidimensional con una forma de 5 x 5. Extrae el valor máximo de la matriz, y los valores máximos de cada uno de sus ejes.

**Primero se crea una matriz de dos dimensiones 'reshape(x,y)' y se le asignan valores dentro de un rango 'arange(z)'**

```python
matriz_2d = np.arange(25).reshape(5,5) #Se crea la matriz de dos dimensiones con forma de 5x5 en un rango de valores dado
```

**Se utiliza la funcion *'max'* para encontrar el valor maximo dentro de la matriz.**

```python
np.max(matriz_2d) #calcula el valor maximo dentro de la matriz
```

**Se utiliza el atributo *'shape'* para obtener que tipo de forma tiene la matriz creada y *'ndim* para confirmar las dimensiones que tiene, y asi verificar que se ha creado segun lo indicado**

```python
matriz_2d.ndim #'ndim' nos especifica la cantidad de dimensiones que tiene la matriz
```

**Se utiliza nuevamente la funcion *'max'* para encontrar el valor maximo, esta vez especificando que el *'axis'* de  busqueda es *'1'* con lo cual se obtiene el valor maximo de cada fila y al especificar el *'axis=0'* obtenemos el valor maximo de cada columna**

```python
print('El maximo de cada fila es', np.max(matriz_2d, axis=1))
print('El maximo de cada columna es', np.max(matriz_2d, axis=0))
```

Muéstrame con ejemplos de diferentes matrices, la regla fundamental de Broadcasting que dice: "las matrices se pueden transmitir / broadcast  si sus dimensiones coinciden o si una de las matrices tiene un tamaño de 1".

**Reutilizado las matrices creadas en el nivel 1, exploramos el funcionamiento de la regla fundamental de *Broadcasting*.**

```python
array_2 =np.array(5) #creacion nuevo array
broad1= np.multiply (array_2, array_numeros) #uso de funcion multiply entre ambos arrays
print(array_numeros) #imprimir array inicial
print(broad1) #imprimir array resultados
```

**La matriz anterior muestra el comportamiento del calculo cuando las matrices tienen diferente tamaño pero una de ellas tiene un tamaño de 1. A continuacion se observa como se comportan cuando la dimension de las dos matrices coincide**
